```python
df_PerYear = pd.concat([df_2017_pivot, df_2018_pivot, df_2019_pivot, df_2020_pivot])
df_PerYear
```

```python
df_PerYear['occupation']= ''
df_PerYear['occupation_code'] = ''
df_PerYear['days_away']=''
df_PerYear['days_away_median'] =''
```

```python
df_PerYear
```

```python
df_PerYear = df_PerYear.rename(columns={'NAICS code' : 'naics_code' , 'Industry':'industry', 'Year':'year',
                                        'Total cases': 'total_cases_category','variable':'age_group','value':'cases' })

df_PerYear
```

```python
df_PerYear.columns
```

```python
df_PerYear_ordenado= df_PerYear[['year', 'occupation', 'occupation_code', 'industry', 
                                 'naics_code','age_group', 'days_away', 'cases',
                                 'total_cases_category','days_away_median']]
df_PerYear_ordenado
```

```python
df_PerYear_ordenado['cases']=pd.to_numeric(df_PerYear_ordenado.cases, errors='coerce')
df_PerYear_ordenado.dtypes
```

```python
df_PerYear_ordenado.to_csv('df_R37.csv', index= False)
```
